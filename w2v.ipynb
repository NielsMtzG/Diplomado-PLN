```python
from gensim.models.keyedvectors import KeyedVectors
```

```python
wordvectors_file_vec = './embeddings-l-model.vec'
cantidad = 100000
```

```python
wordvectors = KeyedVectors.load_word2vec_format(wordvectors_file_vec)
```

```python
print('Similitud entre los términos rey y mujer; definiendo género femenino:')

print(wordvectors.most_similar_cosmul(positive=['rey','mujer'],negative=['hombre']))
```

```python
print('Similitud entre los términos actor y mujer; definiendo género femenino:')
print(wordvectors.most_similar_cosmul(positive=['actor','mujer'],negative=['hombre']))
```

```python
print('Similitud entre los términos hijo y mujer; definiendo género femenino:')
print(wordvectors.most_similar_cosmul(positive=['hijo','mujer'],negative=['hombre']))
```

```python
print('Considerando Conjugaciones:')
print(wordvectors.most_similar_cosmul(positive=['jugar','canta'],negative=['cantar']))
```

```python
print('Considerando Relaciones:')
print(wordvectors.most_similar_cosmul(positive=['habana','chile'],negative=['santiago']))
```

```python
print('Excluyendo al diferente:')
print(wordvectors.doesnt_match(['blanco','azul','rojo','chile']))
```
